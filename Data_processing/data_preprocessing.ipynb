```python
import pandas as pd
from datetime import datetime
```

```python
# url = './conn.log.labeled_longer'
url = "../CTU-IoT-Malware-Capture-60-1/bro/conn.log.labeled"
# START EDITING HERE - DON'T REMOVE THIS COMMENT
df = pd.read_table(url, sep='\s+', 
                   skiprows=8, 
                   engine='python', 
                   names=["ts", "uid", "id.orig_h", 
                          "id.orig_p", "id.resp_h", 
                          "id.resp_p", "proto", "service", 
                          "duration", "orig_bytes", 
                          "resp_bytes", "conn_state", 
                          "local_orig", "local_resp", 
                          "missed_bytes", "history", 
                          "orig_pkts", "orig_ip_bytes", 
                          "resp_pkts", "resp_ip_bytes", 
                          "tunnel_parents","label","detailed-label"],
                   dtype={'ts': float},
#                    nrows=5000000,
                   skipfooter=1
                  )
df = df[df.duration != '-']
df = df[df['id.orig_h'] != '-']
df = df[df['id.resp_h'] != '-']

# def convert_dt(timestamp):
#     print("timestamp = ", timestamp)
    
#     try:
#         return datetime.datetime.fromtimestamp(float(epoch_time))
#     except:
#         return None


# data["Label"] = data.apply(lambda row: preprocess_data(row["Text"]), axis=1)
# df.apply(lambda row: list(encoder(row["Text"]).get('attention_mask')), axis=1)
# df = df[df['orig_bytes'].apply(lambda x: x.isnumeric())]
# df = df[df['resp_bytes'].apply(lambda x: x.isnumeric())]
df = df.astype({'orig_bytes': float, 
                'missed_bytes': float, 
                'resp_bytes': float, 
                'orig_ip_bytes':float,
                'resp_ip_bytes': float
               })
print(df.dtypes)
df['date'] = pd.to_datetime(df['ts'], unit='s')
print('Dimension of df:', df.shape)
display(df.head())
```

```python
sorted_df = df.sort_values(by='date')
sorted_df.tail()
```

```python
sorted_df['Time_diff'] = sorted_df['ts'].diff().fillna(0)
sorted_df.head()
# sorted_df.dtypes
sorted_df['teT'] = sorted_df.date.dt.hour
sorted_df['teS'] = sorted_df.date.dt.hour * 3600 + sorted_df.date.dt.minute * 60 + sorted_df.date.dt.second
```

```python
sorted_df.head(10)
```

```python
sorted_df.info()
sorted_df['total_bytes_overall'] = sorted_df['orig_bytes'] + sorted_df['orig_ip_bytes'] + sorted_df['resp_ip_bytes'] + sorted_df['missed_bytes'] + sorted_df['resp_bytes']
sorted_df['total_packets_overall'] = sorted_df['orig_pkts'] + sorted_df['resp_pkts']
```

```python
final_df = sorted_df.rename(columns = {"id.orig_h": "sa", 
                                       "id.orig_p":"sp",
                                       "id.resp_h": "da",
                                       "id.resp_p": "dp",
                                       "proto": "pr",
                                       "duration": "td",
                                       "service": "stos",
                                       'total_bytes_overall': 'byt',
                                       'total_packets_overall': "pkt",
                                       "date": "te",
                                       "Time_diff": "teDelta"
                                      }) 
```
